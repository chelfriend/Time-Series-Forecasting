```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic time series data
np.random.seed(42)
num_points = 100
time_series = np.random.normal(loc=0, scale=1, size=num_points).cumsum()

# Create features and target
X = np.arange(num_points).reshape(-1, 1)
y = time_series

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict next 10 time points
forecast = model.predict(np.arange(num_points, num_points+10).reshape(-1, 1))

# Plot original time series and forecast
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Original Data')
plt.plot(np.arange(num_points, num_points+10), forecast, label='Forecast', color='red')
plt.title('Linear Regression Forecast')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
