```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Generate synthetic time series data
np.random.seed(42)
num_points = 100
time_series = np.random.normal(loc=0, scale=1, size=num_points).cumsum()

# Fit ARIMA model
model = ARIMA(time_series, order=(1,1,1))
model_fit = model.fit()

# Forecast next 10 time points
forecast = model_fit.forecast(steps=10)

# Plot original time series and forecast
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Original Data')
plt.plot(np.arange(num_points, num_points+10), forecast, label='Forecast', color='red')
plt.title('ARIMA Forecast')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
